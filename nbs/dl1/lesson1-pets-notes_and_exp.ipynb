In general data exists in many forms and while there are general patterns for how the data is made availabe, the data will have to be manipulated and 'cleaned up' before we can use it. In this particular dataset, labels are stored in the filenames themselves. We will need to extract them to be able to classify the images into the correct categories. Fortunately, the fastai library has a handy function made exactly for this, `ImageDataBunch.from_name_re` gets the labels from the filenames using a [regular expression](https://docs.python.org/3.6/library/re.html).

```python
fnames = get_image_files(path_img)
fnames[:5]
```

Now we need to sort out the picctures, but all of them are in tthe same folder. The only distinction is just the file name tha embeds the category. There is a function in fast.ai that can help with this but and it gets passade as a parameter a regular expression that is used to sort the categories. This is what is stored in the variable `pat` below. A good explanation that unpacks the regular expression is here: https://medium.com/@youknowjamest/parsing-file-names-using-regular-expressions-3e85d64deb69. But basically this is what the expression below means (ready from right to left):


`|`$`|end of search|
|`.jpg`|last chars to be found in the search string, also right file format checking|
|`\d`|numerical digits, ‘+’ sign denotes can be one or more of them|
|`_`|should come before the start of digits|
|`()`|denotes a group of characters|
|`[]`|denotes another subgroup if characters in the previous group|
|`^/`|‘^’ denotes negation, so ‘+’ says all other chars except ‘/’|
|`( [ ^/ ] + )`|searches all characters except ‘/’|
|`/`|first ‘/’ in regex says, end of search|
|`r`|The string should be a raw string. Otherwise, `\d` would have to be written as `\\d` so that Python doesn’t interpret it to be a special character.|`



```python
np.random.seed(2)
pat = r'/([^/]+)_\d+.jpg$'
```

Next we load the data using the above regular expression.  

```python
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(), size=224, bs=bs
                                  ).normalize(imagenet_stats)
```

A few important things are happening here (read the documentation for `class ImageBunch` to get more details):

* The images are resized to 224x224. Why this size? Because this is the size that the ResNet34 architecture was trained on. If we use a different architecture, this size will have to be changed.
* Images get centered, cropped and zoomed by the transformation functions
* Data is normalized. For an imageP (in this case), the image is composed of 3 channels: Red (R), Green (G) and Blue (B). Each pixel ranges in value from (integers) 0 to 255. Because in real life (and this is my personal experience from my photography work) all of the channels have different ranges of values (min an max) in the 0 - 255 valid range. Models like to have data presented to them where all of the channles have a mean of 0 and a standard deviation of 1 - the normalizatin function takes care of that (ref: https://stats.stackexchange.com/questions/211436/why-normalize-images-by-subtracting-datasets-image-mean-instead-of-the-current)

**Worth noting** If there are problems training a model, a quick thing that should be chcked is that the data has been normalized properly. 

Note the `normalize(imagenet_stats)` statement. This normalizes the data using ImageNet data since that is what the model we are using was originally trained on.

in fast.ai a model gets creted by a `learner`. To get this going you call `cnn_learner` and this creates the learner. It gets told about the data and loads the weights of the ResNet34 architure. We also tell it the metrics we want to see.

```python
# Run this cell to get a quick peek at a summary of what this function does
# and also to get a link to the fast.ai API documentation
doc(cnn_learner)
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

The fast.ai function `learn.fit_one_cyle()` takes as a  paramenter an integer that says how many times we want to iterate through the entire dataset. Each pass through the data it gets better and better. Right now other than the number of iterations, we use defaults.

```python
learn.fit_one_cycle(4)  
```

## Analysis of the Results

We have trained a model and saved it for later use, now we can see how it performed. 

We will first see which were the categories that the model most confused with one another. We will try to see if what the model predicted was reasonable or not. In this case the mistakes look reasonable (none of the mistakes seems obviously naive). This is an indicator that our classifier is working correctly. 

Furthermore, when we plot the confusion matrix, we can see that the distribution is heavily skewed: the model makes the same mistakes over and over again but it rarely confuses other categories. This suggests that it just finds it difficult to distinguish some specific categories between each other; this is normal behaviour. The `ClassificationInterpretation` class has methods for creating the confusion matrix and plotting missclassified images. It takes as its input parameter the `learner` that we careated earlier.

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

So, up to this point we have a pretty good model. Is there anything we can do (at this stage of things)to make the model better?

**Yes** We can fine-tune our model. When we earlier run the training code, what actually happened under the hood was somethig called `transfer learning`. In short what really happened is that we added a few new layersm, froze the other part of the network and only trained the new layers. This made for very quick training passes.

To get a model that (should be) will be better we really need to train all of the layers and to do that, we need to tell the fast.ai library to unfreeze all of the layers by calling `learn.unfreeze()`
